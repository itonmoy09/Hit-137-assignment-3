```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Screen dimensions
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# Fonts
FONT = pygame.font.SysFont("Arial", 20)

# Load Sounds
shoot_sound = pygame.mixer.Sound("C:\\CDU 1st semester\\Software Now\\Question 2\\shoot_sound.WAV")
player_hit_sound = pygame.mixer.Sound("C:\\CDU 1st semester\\Software Now\\Question 2\\player_hit_sound.WAV")
game_over_sound = pygame.mixer.Sound("C:\\CDU 1st semester\\Software Now\\Question 2\\game_over_sound.WAV")
enemy_hit_sound = pygame.mixer.Sound("C:\\CDU 1st semester\\Software Now\\Question 2\\enemy_hit_sound.WAV")

# Load Images
player_image = pygame.image.load("C:\\CDU 1st semester\\Software Now\\Question 2\\Player_1.jpg")
enemy_image = pygame.image.load("C:\\CDU 1st semester\\Software Now\\Question 2\\Enemy_1.jpg")
boss_image = pygame.image.load("C:\\CDU 1st semester\\Software Now\\Question 2\\Boss.jpg")
extra_life_image = pygame.image.load("C:\\CDU 1st semester\\Software Now\\Question 2\\extra_life.jpg")

# Player class
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.transform.scale(player_image, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.x = 100
        self.rect.y = SCREEN_HEIGHT - 100
        self.speed = 5
        self.jump_speed = 10
        self.gravity = 0.5
        self.velocity_y = 0
        self.health = 100
        self.lives = 3
        self.score = 0
#Player movement to set its moving right, left and also jump
    def update(self):
        self.velocity_y += self.gravity
        self.rect.y += self.velocity_y

        if self.rect.bottom > SCREEN_HEIGHT:
            self.rect.bottom = SCREEN_HEIGHT
            self.velocity_y = 0

    def move_left(self):
        self.rect.x -= self.speed

    def move_right(self):
        self.rect.x += self.speed

    def jump(self):
        if self.rect.bottom == SCREEN_HEIGHT:
            self.velocity_y = -self.jump_speed

    def shoot(self, projectiles):
        shoot_sound.play()
        projectile = Projectile(self.rect.centerx, self.rect.top)
        projectiles.add(projectile)

# Projectile class
class Projectile(pygame.sprite.Sprite):
    def __init__(self, x, y):
        super().__init__()
        self.image = pygame.Surface([10, 20])
        self.image.fill(RED)
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.speed = 10
        self.damage = 25

    def update(self):
        self.rect.y -= self.speed
        if self.rect.bottom < 0:
            self.kill()

# Enemy class
class Enemy(pygame.sprite.Sprite):
    def __init__(self, x, y):
        super().__init__()
        self.image = pygame.transform.scale(enemy_image, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.speed = 2
        self.health = 50

    def update(self):
        self.rect.x -= self.speed
        if self.rect.right < 0:
            self.kill()
#enemy can kill the player if enemy hit the player
    def hit(self, damage):
        self.health -= damage
        if self.health <= 0:
            enemy_hit_sound.play()
            self.kill()

# Boss class to get boss in the game and make boss health more then the normal enemy
class Boss(pygame.sprite.Sprite):
    def __init__(self, x, y):
        super().__init__()
        self.image = pygame.transform.scale(boss_image, (100, 100))
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.speed = 1
        self.health = 200

    def update(self):
        self.rect.x -= self.speed
        if self.rect.right < 0:
            self.kill()
#boss can kill the player by hiting the player by it's body
    def hit(self, damage):
        self.health -= damage
        if self.health <= 0:
            enemy_hit_sound.play()
            self.kill()

# Collectible class
class Collectible(pygame.sprite.Sprite):
    def __init__(self, x, y, type):
        super().__init__()
        if type == 'health':
            self.image = pygame.Surface([30, 30])
            self.image.fill(GREEN)
        elif type == 'life':
            self.image = pygame.transform.scale(extra_life_image, (30, 30))
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.type = type

# Game over function
def game_over_screen(screen):
    game_over_sound.play()
    screen.fill(BLACK)
    game_over_text = FONT.render("Game Over! Press R to Restart or Q to Quit", True, WHITE)
    screen.blit(game_over_text, (SCREEN_WIDTH // 2 - game_over_text.get_width() // 2, SCREEN_HEIGHT // 2))
    
    # Restart button as the question asked 
    restart_button = pygame.Rect(SCREEN_WIDTH // 2 - 50, SCREEN_HEIGHT // 2 + 50, 100, 50)
    pygame.draw.rect(screen, WHITE, restart_button)
    restart_text = FONT.render("Restart", True, BLACK)
    screen.blit(restart_text, (restart_button.x + 10, restart_button.y + 10))
    
    pygame.display.flip()

    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_r:
                    waiting = False
                    main()
                if event.key == pygame.K_q:
                    pygame.quit()
                    exit()
            if event.type == pygame.MOUSEBUTTONDOWN:
                if restart_button.collidepoint(event.pos):
                    waiting = False
                    main()

# Main game function
def main():
    screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    pygame.display.set_caption("Side-Scrolling 2D Game")

    player = Player()
    projectiles = pygame.sprite.Group()
    enemies = pygame.sprite.Group()
    bosses = pygame.sprite.Group()
    collectibles = pygame.sprite.Group()

    all_sprites = pygame.sprite.Group()
    all_sprites.add(player)

    # Initial enemy and collectible spawn
    for _ in range(5):
        enemy = Enemy(random.randint(SCREEN_WIDTH, SCREEN_WIDTH + 1000), random.randint(100, SCREEN_HEIGHT - 100))
        enemies.add(enemy)
        all_sprites.add(enemy)

    for _ in range(2):
        collectible = Collectible(random.randint(SCREEN_WIDTH, SCREEN_WIDTH + 1000), random.randint(100, SCREEN_HEIGHT - 100), 'health')
        collectibles.add(collectible)
        all_sprites.add(collectible)

    for _ in range(1):
        collectible = Collectible(random.randint(SCREEN_WIDTH, SCREEN_WIDTH + 1000), random.randint(100, SCREEN_HEIGHT - 100), 'life')
        collectibles.add(collectible)
        all_sprites.add(collectible)

    # Timer for spawning new enemies and bosses
    enemy_spawn_time = 3000  # Spawn an enemy every 3000 milliseconds (3 seconds)
    boss_spawn_time = 15000  # Spawn a boss every 15000 milliseconds (15 seconds)
    last_enemy_spawn = pygame.time.get_ticks()
    last_boss_spawn = pygame.time.get_ticks()

    clock = pygame.time.Clock()
    running = True

    while running:
        current_time = pygame.time.get_ticks()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    player.move_left()
                if event.key == pygame.K_RIGHT:
                    player.move_right()
                if event.key == pygame.K_UP:
                    player.jump()
                if event.key == pygame.K_SPACE:
                    player.shoot(projectiles)
                    all_sprites.add(projectiles)

      
        # Check if it's time to spawn a new enemy
        if current_time - last_enemy_spawn > enemy_spawn_time:
            enemy = Enemy(random.randint(SCREEN_WIDTH, SCREEN_WIDTH + 1000), random.randint(100, SCREEN_HEIGHT - 100))
            enemies.add(enemy)
            all_sprites.add(enemy)
            last_enemy_spawn = current_time

        # Check if it's time to spawn a new boss
        if current_time - last_boss_spawn > boss_spawn_time:
            boss = Boss(random.randint(SCREEN_WIDTH, SCREEN_WIDTH + 1000), random.randint(100, SCREEN_HEIGHT - 100))
            bosses.add(boss)
            all_sprites.add(boss)
            last_boss_spawn = current_time

        all_sprites.update()

        for projectile in projectiles:
            enemy_hit_list = pygame.sprite.spritecollide(projectile, enemies, False)
            for enemy in enemy_hit_list:
                enemy.hit(projectile.damage)
                projectile.kill()
                player.score += 10

            boss_hit_list = pygame.sprite.spritecollide(projectile, bosses, False)
            for boss in boss_hit_list:
                boss.hit(projectile.damage)
                projectile.kill()
                player.score += 50

        for enemy in enemies:
            if pygame.sprite.collide_rect(player, enemy):
                player_hit_sound.play()
                player.lives -= 1
                if player.lives <= 0:
                    game_over_screen(screen)
                    running = False

        for boss in bosses:
            if pygame.sprite.collide_rect(player, boss):
                player_hit_sound.play()
                player.lives -= 1
                if player.lives <= 0:
                    game_over_screen(screen)
                    running = False

        for collectible in collectibles:
            if pygame.sprite.collide_rect(player, collectible):
                if collectible.type == 'health':
                    player.health = min(player.health + 25, 100)
                elif collectible.type == 'life':
                    player.lives += 1
                collectible.kill()
                player.score += 5

        screen.fill(BLACK)
        all_sprites.draw(screen)
#front to show on the game
        health_text = FONT.render(f"Health: {player.health}", True, WHITE)
        lives_text = FONT.render(f"Lives: {player.lives}", True, WHITE)
        score_text = FONT.render(f"Score: {player.score}", True, WHITE)

        screen.blit(health_text, (10, 10))
        screen.blit(lives_text, (10, 30))
        screen.blit(score_text, (10, 50))

        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

if __name__ == "__main__":
    main()
```
